```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
URL='https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv'
df=pd.read_csv(URL)
display(df.head)
df.shape[0]
df.shape[1]
print(f"shape:{df.shape}  ({df.shape[0]}) rows*{df.shape[1]}columns")


```

```python
df.describe()
```

```python
df.info()
```

```python
#data sets
for col in ['age','fare']:
  q1,q3=df[col].quantile(0.25),df[col].quantile(0.75)
  print(f"\n{'='*401}\n{col.upper()}")
  print(f"mean :{df[col].mean():.2f}")
  print(f"median :{df[col].median():.2f}")
  print(f"std dev:{df[col].std():.2f}")
  print(f"max/min :{df[col].min():.2f}/{df[col].max():.2f}")
  print(f"q1/q3:{q1:.2f}/{q3:.2f}|IQR:{q3-q1:.2f}")


print(f"\nmode of 'embarked':{df['embarked'].mode()[0]}")
print(f"mode of 'age':{df['age'].mode()[0]}")




```

```python

# Missing value summary
missing = pd.DataFrame({
    'count': df.isnull().sum(),
    'percent (%)': (df.isnull().sum() / len(df) * 100).round(1)
})

print("--- Missing Values ---")
print(missing[missing['count'] > 0].sort_values('percent (%)', ascending=False))

# Duplicate rows
print(f"\nDuplicate rows: {df.duplicated().sum()}")

# Unique counts
print("\n--- Unique Values Per Column ---")
print(df.nunique().sort_values())
```
